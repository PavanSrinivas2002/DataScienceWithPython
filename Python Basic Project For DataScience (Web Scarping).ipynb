```python
#Web Scraping with Data frames
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/World_population"
```

```python
data=requests.get(url).text
```

```python
soup=BeautifulSoup(data,"html5lib")
```

```python
tables = soup.find_all('table') 
len(tables)
```

```python
for index,table in enumerate(tables):
    if ("Most populous country" in str(table)):
        table_index = index
print(table_index)
```

```python
#print(tables[table_index])
population_data = pd.DataFrame(columns=["Continent", "Density", "Population", "Most populous country", "Most populous city"])
```

```python
rows=tables[1].find_all('tr')
```

```python
for row in rows:
    col=row.find_all('td')
    if(col!=[]):
        Continent=col[0].text.strip()
        Density=col[1].text.strip()
        Population=col[2].text.strip()
        Mostpopulouscountry=col[3].text.strip()
        Mostpopulouscity=col[4].text.strip()
        population_data=population_data.append({"Continent":Continent, "Density":Density, "Population":Population, "Most populous country":Mostpopulouscountry, "Most populous city":Mostpopulouscity},
                                               ignore_index=True)
```

```python
population_data
```
